# 【確認課題】Matplotlib

```python
# 使うライブラリ
# !pip install japanize-matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()

hist_data = pd.read_csv('data/hist_data.csv')
line_data = pd.read_csv('data/line_data.csv')
scatter_data = pd.read_csv('data/scatter_data.csv')
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
plt.title('aデータとbデータの温度に関する度数分布', fontsize=13)
plt.xlabel('温度(℃)', fontsize=10)
plt.ylabel('度数', fontsize=10)

plt.hist(hist_data['aデータ'], bins=30, histtype="stepfilled", color='darkorange', alpha=0.5, label='aデータ')
plt.hist(hist_data['bデータ'], bins=30, histtype="stepfilled", color='darkturquoise', alpha=0.5, label='bデータ')
plt.legend(loc='upper right', fontsize=10)
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
plt.title('クラス０～２のlengthとwidthに関する散布図', fontsize=13)
plt.xlabel('length (cm) ', fontsize=10)
plt.ylabel('width (cm) ', fontsize=10)
plt.grid(True)

df_class0 = scatter_data[scatter_data['class'] == 0]
df_class1 = scatter_data[scatter_data['class'] == 1]
df_class2 = scatter_data[scatter_data['class'] == 2]

plt.scatter(df_class0['length (cm)'], df_class0['width (cm)'], c='r', label='class0')
plt.scatter(df_class1['length (cm)'], df_class1['width (cm)'], c='b', label='class1')
plt.scatter(df_class2['length (cm)'], df_class2['width (cm)'], c='g', label='class2')
plt.legend(loc='upper right', fontsize=10)
```

```python
plt.figure()

plt.title('期間ごとの件数の推移', fontsize=13)
plt.xlabel('期間', fontsize=10)
plt.ylabel('件数', fontsize=10)
plt.grid(True)
plt.xticks(rotation=30)

plt.plot(line_data['期間'], line_data['件数'], marker='o')


plt.figure()

plt.title('期間ごとの費用の推移', fontsize=13)
plt.xlabel('期間', fontsize=10)
plt.ylabel('費用(円)', fontsize=10)
plt.grid(True)
plt.xticks(rotation=30)

plt.bar(line_data['期間'], line_data['費用'])
```
